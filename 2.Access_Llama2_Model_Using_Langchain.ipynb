```python
#3.Import All the Required Libraries
from langchain.llms import HuggingFacePipeline
#with the help of hugging dace pipeline we will be able to load my large language model which is Llama2
```

```python
#Import tokenizer - text to vectors
from transformers import AutoTokenizer
```

```python
import transformers
```

```python
import torch
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#4.Load the Llama Model
model="meta-llama/Llama-2-7b-chat-hf" #official model
# model="daryl149/llama-2-7b-chat-hf" #public repository cloned version of the model
```

```python
tokenizer=AutoTokenizer.from_pretrained(model)
```

```python
#create pipeline
pipeline=transformers.pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    device_map="auto",
    max_length=1000,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id
    )
```

```python
llm=HuggingFacePipeline(pipeline=pipeline, model_kwargs={'temperature':0})
```

```python
prompt="What would be a good name for a company that makes car"
```

```python
print(llm(prompt))
```

```python
prompt="I want to open a restaurant for south indian food. Suggest me a fence name for this"
```

```python
print(llm(prompt))
```

```python
#5.Prompt Templates - create a custom prompt
#Import All the Required Libraries
from langchain.prompts import PromptTemplate
```

```python
from langchain.chains import LLMChain
```

```python
#Example1
prompt_template1=PromptTemplate(input_variables=["cuisine"],
                               template="I want to open a restaurant for {cuisine} food. Suggest a fency name for this")
```

```python
input_prompt=prompt_template1.format(cuisine="south indian")
```

```python
print(input_prompt)
```

```python
#Example 2
prompt_template2=PromptTemplate(input_variables=["book_name"],
                               template="Provide me a concise summary of the book {book_name}")
```

```python
input_prompt=prompt_template2.format(book_name="Rich dad poor dad")
```

```python
print(input_prompt)
```

```python
chain = LLMChain(llm=llm, prompt=prompt_template2, verbose=True)
response= chain.run("Rich dad Poor dad")
print(response)
```

```python
chain = LLMChain(llm=llm, prompt=prompt_template1, verbose=True)
response= chain.run("South indian")
print(response)
```
